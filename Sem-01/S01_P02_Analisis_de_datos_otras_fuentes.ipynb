### **B.	Estadísticas descriptivas**
a.	Cree un set de datos  con 20 datos que contenga al menos dos atributos continuos y una variable categórica (por ejemplo: nombre, nota, edad).

```python
import pandas as pd
d = {'Nombre' : pd.Series(['Diego', 'Victor',	'Alex' ,'Darsi' ,'Karen','Pablo','Alejandro','Maritza','Vivian','Nicole','Yoshelin','Ana','Juan','Carlos','Karina','Derling','Roger','Miguel','Joaquin','Efrain' ]),
      'Nota' : pd.Series([91,74,66,96,49,43,64,32,64,71,36,65,37,45,94,60,83,45,92,49]),
      'Edad' :  pd.Series([28,35,31,25,21,26,25,31,30,27,33,22,20,21,29,23,27,21,28,27])}


```

b.	Transforme dicho set de datos  a un data frame de pandas.

```python
df_datos = pd.DataFrame(d)
df_datos
```

```python
df_datos.dtypes
```

c.	Obtenga estadísticas descriptivas de tendencia central.

```python
# Obtenemos las estadísticas del Campo Nota
print('Las estadísticas descriptivas de tendencia central respecto a las notas tenemos:')
print('promedio: ',df_datos['Nota'].mean())
print('mediana: ',df_datos['Nota'].median())
print('moda: ', df_datos['Nota'].mode().tolist())

```

```python
# Obtenemos las estadísticas del Campo Edad
print('Las estadísticas descriptivas de tendencia central respecto a la edad tenemos:')
print('promedio: ',df_datos['Edad'].mean())
print('mediana: ',df_datos['Edad'].median())
print('moda: ', df_datos['Edad'].mode().tolist())
```

d.	Obtenga estadísticas descriptivas de dispersión.

```python
import numpy as np
print('Las estadísticas descriptivas de dispersión respecto a las notas tenemos:')
print('rango: ', np.round(df_datos['Nota'].max()-df_datos['Nota'].min(),2))
print('Varianza: ',np.round(df_datos['Nota'].var(),2))
std = df_datos['Nota'].std()
print('Desviación estándar: ', np.round(std,2))
print('Coeficiente de variación: ', np.round(std/df_datos['Nota'].mean(),2))
```

```python
print('Las estadísticas descriptivas des dispersión respecto a la edad tenemos:')
print('rango: ', np.round(df_datos['Edad'].max()-df_datos['Edad'].min(),2))
print('Varianza: ',np.round(df_datos['Edad'].var(),2))
std = df_datos['Edad'].std()
print('Desviación estándar: ', np.round(std,2))
print('Coeficiente de variación: ', np.round(std/df_datos['Edad'].mean(),2))
```
